# Data Cleaning

```python
book.loc[206:"Date of Publications"].head()
```

```python
book["Date of Publication"]
```

```python
# Extracting the Date of publication column data using regex 
book["Date of Publication"] = book["Date of Publication"].str.extract(r'^(\d{4})',expand = False)
book["Date of Publication"]
```

```python
#Identifing the unique values of Date of publication column
book["Date of Publication"].unique()
```

```python
# Calculating the null values in Date of Publication column
book["Date of Publication"].isnull().sum()
```

```python
# Finding the most frequently using value in Date of Publication column
book["Date of Publication"].mode()
```

```python
# Filling the NaN value as Most frequent value in Date of Publication column
book["Date of Publication"] = book["Date of Publication"].fillna(value= "1879")
```

```python
book["Date of Publication"].isnull().sum()
```

```python
book["Date of Publication"].unique
```

```python
book.isnull().sum()
```

```python
book.info()
```

```python
# Converting Date of Publication Values Object to Numeric
book["Date of Publication"] = pd.to_numeric(book["Date of Publication"]) 
```

```python
book["Date of Publication"]
```

```python
# Calculating the null values in Date of Publication column
book["Publisher"].isnull().sum()
```

```python
# Filling the NaN value as not none in Publisher Column
book["Publisher"] = book["Publisher"].fillna(value = "not none") 
```

```python
book.isnull().sum()
```

```python
# Filling the NaN value as not none in Author Column
book["Author"] = book["Author"].fillna(value = "not none")
```

```python
book["Author"].isnull()
```

```python
book.info()
```

```python
# Examine the place of publication data
book["Place of Publication"]
```

```python
# Finding the most frequently using value in Place of Publication column
book["Place of Publication"].mode()
```

```python
# Calculating the most frequently value in Place of Publication column
book["Place of Publication"].str.contains("London").sum()
```

```python
# Calculating Oxford value in Place of publication column
book["Place of Publication"].str.contains("Oxford").sum()
```

```python
# Calculating all values sum of Place of Publication column
book["Place of Publication"].value_counts()
```

```python
# Changing the and doing correct formating of Place of publication column because its contian some unusals elements

book['Place of Publication'] = np.where(book["Place of Publication"].str.contains("London"), 'London',
                                      np.where(book["Place of Publication"].str.contains("Oxford"), 'Oxford',
                                               book["Place of Publication"].str.replace('-', ' ')))
```

```python
book["Place of Publication"].isnull()
```
